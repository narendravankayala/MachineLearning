# II. Linear Algebra and Probability Theory

## A. Linear Algebra
* **Scalars** : Sclar is a single number
* **Vectors** : Vector is an array of numbers and each element can be accessed by the index. Vectors are usually represented in lowercase letters
* **Matrix**  : Matrix is an array of 2D numbers and each element can be accessed by two indices. Matrices are usually represented in upper case letters
* **Tensors** : Tensor is the representation of numbers on more than 2 axes. We denote a tensor named “A”with this typeface:A. We identify the element of **_A_** at coordinates (i, j, k) by writing **A**(i,j,k)
* **Transpose** : Transpose of a matrix is the mirror image of the matrix across diagonal line. 
* **Multiplying Matrices and Vectors** : The matrix productof matrices A and B is a third matrix C. Inorder for this product to be deﬁned,A must have the same number of columns as B has rows.  **C =AB**
* **Identity Matrix** :An identity matrix is a matrix that does not change any vector when wemultiply that vector by that matrix.The structure of the identity matrix is simple: all the entries along the maindiagonal are 1, while all the other entries are zero.
A_{n,n} = 
 \begin{pmatrix}
  1 & 0 & \cdots & 0 \\
  0 & 1 & \cdots & 0 \\
  \vdots  & \vdots  & \ddots & \vdots  \\
  0 & 0 & \cdots & 1
 \end{pmatrix}
* **Linear Dependence** :Formally, a linear combination of some set of vectors {v(1), . . . , v(n)} is given by multiplying each vector v(i) by a corresponding scalar coefficient and adding the results
* **Span** : The span of a set of vectors is the set of all points obtainable by linear combination of the original vectors.
* **Norms** : On an intuitive level, the norm of a vector **_x_** measures the distance fromthe origin to the point **_x_**.
* **Orthogonal matrix** : An orthogonal matrix is a square matrix whose rows are mutually orthonormal and whose columns are mutually orthonormal, which means the transpose the matrix is equal to inverse of the matrix
* **Eigendecomposition** : An eigenvector of a square matrix A is a nonzero vector v such that multiplication by A alters only the scale of v: Av = λv. The scalar λ is known as the eigen value corresponding to this eigen vector
* **Principal Components Analysis** : Suppose we have a collection of m points {x(1), . . . , x(m)} and we want to apply lossy compression to these points. Lossy compression means storing the points in a way that requires less memory but may lose some precision. We wantto lose as little precision as possible.We can encode these points is to represent a lower-dimensional version of them.

## B. Probability Theory
* **Random Variables** : Arandom variableis a variable that can take on diﬀerent values randomly. Wetypically denote the random variable itself with a lowercase letter in plain typeface,and the values it can take on with lowercase script letters.
* **Probability Distributions** : A probability distributionis a description of how likely a random variable or set of random variables is to take on each of its possible states. 
* **Discrete Variables and Probability Mass Functions** : A probability distribution over discrete variables may be described using aproba-bility mass function(PMF).
* **Marginal Probability** : The probability distribution over the subset is known as the marginal probability distribution$$p(x) = \int p(x,y)dy$$
* **Conditional Probability** : Probability of some event, given that someother event has happened$$P(Y=y | X=x) =\frac {P(Y=y,X =x)}{P(X=x)}$$
* **Bernoulli Distribution** : The Bernoulli distributionis a distribution over a single binary random variable.It is controlled by a single parameter φ ∈[0,1], which gives the probability of therandom variable being equal to 1.
* **Multinoulli Distribution** : The multinoulli distribution is a distribution over a single discrete variable with k different states, where k is finite. This is also called as categorical distribution.
* **Bayes’ Rule** : The probability of an event, based on prior knowledge of conditions that might be related to the event.$$P(x|y) =\frac {P(x)P(y|x)}{P(y)}$$
